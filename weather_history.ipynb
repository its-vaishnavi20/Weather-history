# Separting the target_summary variable from the remaining dataset

```python
target_summary = data['Summary']
data.drop(columns='Summary', inplace = True)
```

**one hot encoding for all the variables with categorical data**

```python
data_dummies=pd.get_dummies(data)
```

```python
data_dummies.shape
```

```python
data_dummies.head()
```

**Shuffling the Data before splitting it**

```python
from sklearn.utils import shuffle
df_shuffled = shuffle(data_dummies, random_state = 42)
```

# **Temperature is the dependent variable that is to be predicted**
##**Separating the "Temperature" variable, which is to be used as the target variable**

```python
DV = 'Temperature (C)'
X = df_shuffled.drop(DV,axis=1)
```

```python
Y = df_shuffled[DV]
```

**Splitting x and y into train and test set**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```



#1.Simple  linear regression with Humidity as the independent variable

##Fitting the model to Humidity column for performing simple single regression




```python
model.fit(X_train[['Humidity']],y_train)
```

```python
LinearRegression(copy_X=True,fit_intercept=True, n_jobs=None, normalize=False)
```

```python
intercept =model.intercept_
```

```python
print(intercept)
```

**To print the coefficients of the model , use the following code:**

```python
coefficient = model.coef_
```

```python
coefficient
```

**GENERATE PREDICTS**

```python
predictions = model.predict(X_test[['Humidity']])
predictions_X_train = model.predict(X_train[['Humidity']]) #for plotting the regression line
```

**Visualizing sepearation of training and Testing data points**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.plot(X_train[['Humidity']],y_train,'o',color = 'black',label='Training dataset')
plt.plot(X_test[['Humidity']],y_test,'o',color = 'red', label='Testing Dataset')
plt.legend(loc='upper left')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.title('Splitting of Training and Testing dataset',fontsize=22)
plt.show()
```

**Generating plots for regression**

```python
plt.figure(figsize=(15,7))
plt.plot(X_train[['Humidity']],y_train,'o',color = 'black',label='Actual Value')
plt.plot(X_train[['Humidity']],predictions_X_train,'--r',label='Predicted Value')
plt.legend(loc='upper left')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.title('Temperature Variation in accordance with Humidity for Training Data',fontsize=22)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(X_test[['Humidity']],y_test,'o',color = 'black',label='Actual Test Value')
plt.plot(X_test[['Humidity']],predictions,'--r',label='Predicted Value')
plt.legend(loc='upper left')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.title('Temperature Variation in accordance with Humidity for Test Data',fontsize=22)
plt.show()
```
